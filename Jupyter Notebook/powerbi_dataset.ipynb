```python
import pandas as pd 
```

```python
df = pd.read_csv('vietnam_housing_dataset.csv')

df.head()
```

```python
df = df.drop(columns=['House direction', 'Balcony direction'])

df.head()
```

```python
df['Address'] = df['Address'].str.split(', ').str[-1].str.strip().str.split('.').str.get(0)

df['Address'].unique()
```

```python
import unicodedata

df['Address'] = df['Address'].apply(lambda x: unicodedata.normalize('NFC', x))
```

```python
df['Address'] = df['Address'].apply(lambda x: 'Hà Nội' if x in ['HN', 'Quận Nam Từ Liêm', 'Hai Bà Trưng,']
                                               else 'Hồ Chí Minh' if x in ['Hồ Chí Minh giá 2tỷ380', 'Đường số 11', 'Quận Bình Thạnh',
                                                                     'TPHCM', 'TpHCM', 'Phòng công chứng Nguyễn Thị Thành',
                                                                     'Bán nhà chính chủ Phó Đức Chính khu Bà Chiểu - trung tâm Bình Thạnh giá cực tốt',
                                                                     'Hồ Chí Mính', 'Quận 8', 'TP Hồ Chí Minh']
                                               else x)
```

```python
df['Address'] = df['Address'].apply(lambda x: 'Hanoi' if 'Hà Nội' in x else 'Ho Chi Minh City' if 'Hồ Chí Minh' in x else 'Others')
df['Address'].unique()
```

```python
df = df.rename(columns={'Address': 'City'})

df.head(20)
```

```python
df.to_csv('powerbi_dataset.csv', index=False)
```
